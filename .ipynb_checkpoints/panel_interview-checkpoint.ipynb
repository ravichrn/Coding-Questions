```python
def parse_dataset(dataset):
    res = []
    curr_uuid = None
    curr_seq_num = None
    for data in dataset:
        if data['uuid'] != curr_uuid:
            res.append({'uuid':data['uuid'], 'file_paths':[data['file_name']]})
            curr_uuid = data['uuid']
        else:
            res[-1]['file_paths'].append(data['file_name'])
    return res
```

```python
class Sample:
    def __init__(self, value):
        self.value = value

#assuming a list of samples, get the indices of start and end of consecutive stream of samples greater than 100, as list of tuples
def parse_samples(sample_list):
    if sample_list:
        res = []
        start_ind = None
        for index, data in enumerate(sample_list):
            if data.value > 100:
                if start_ind is None: 
                    start_ind = index
            else: 
                if start_ind is not None:
                    end_ind = index-1
                    res.append((start_ind, end_ind))
                    start_ind = None
        if start_ind is not None:
            res.append((start_ind, len(sample_list)-1))
        return res
                        
    return 0
```

```python
sample_list = [Sample(4), Sample(6), Sample(101), Sample(156), Sample(345), Sample(567), Sample(76), Sample(67), Sample(987), Sample(546), Sample(988), Sample(434)]
```

```python
parse_samples(sample_list)
```
