```python
#move nulls/zeros to end

lst = [0,1,2,3,0,0,4,5,0,0,7,8]

def move_zeroes(numbers):
    temp = 0
    for i in range(len(numbers)):
        if numbers[i] != 0:
            numbers[i],numbers[temp] = numbers[temp],numbers[i]
            temp+=1
    return numbers
        
```

```python
print(move_zeroes(lst))
```

```python
# fill null values
lst = [None, 1,2,3,None, None,4,5,None,None,7,8]

def fill_null(num):
    temp = None
    for i in range(len(num)):
        if not num[i]:
            num[i] = temp
        else: temp = num[i]
    return num
```

```python
print(fill_null(lst))
```

```python
num = 1234567781990

def largest_num(num):
    lst = [int(i) for i in str(num)]
    lst.sort(reverse=True)
    nu = ''.join([str(i) for i in lst])
    return int(nu)

def largest_odd_val_num(num):
    lst = [int(i) for i in str(num) if int(i)%2 != 0]
    lst.sort(reverse=True)
    nu = ''.join([str(i) for i in lst])
    return int(nu)
```
